```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error,accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
```

```python

df=pd.read_csv("Breast_Cancer.csv")


print(df.columns)
print(df.isnull().sum())
print(df.shape)
print(df.dtypes)
print(df.info)
df.columns = df.columns.str.strip()

```

```python
#we define x and y x for what we have info about it and y for what we want to predict 
x=pd.get_dummies(df[["area","bedrooms", "bathrooms", "prefarea"]], drop_first=True)
y=df["price"]

#then we should train the x,y and split the train for 80% of the info then 20% for test 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#then we want to get which algorthem we want to use
model=LinearRegression()
model.fit(x_train,y_train)
#predict what we want
predict=model.predict(x_test)

mse=mean_squared_error(y_test,predict)
r2=r2_score(y_test,predict)
print("mse: ",mse)
print("r2: ",r2)
```

```python
df=pd.read_csv("Breast_Cancer.csv")
df.columns=df.columns.str.strip()
label_cols=["T Stage", "N Stage", "6th Stage", "differentiate", "Grade", 
              "A Stage", "Estrogen Status", "Progesterone Status", "Status"]

le=LabelEncoder()
for col in label_cols:
    df[col]=le.fit_transform(df[col])



df=pd.get_dummies(df,columns=["Race", "Marital Status"],drop_first=True)

x=df[["T Stage", "N Stage", "6th Stage", "differentiate", "Grade", 
              "A Stage", "Estrogen Status", "Progesterone Status"]]
y=df["Status"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)

acuracy=accuracy_score(y_test,y_pred)

print("acuracy:",acuracy)
```
