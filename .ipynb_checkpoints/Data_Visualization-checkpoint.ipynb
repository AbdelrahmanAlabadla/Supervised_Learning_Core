```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Histogram graph


ages=[22, 25, 25, 30, 30, 30, 35, 40, 40, 45]

plt.hist(ages,bins=5,color="black",edgecolor="white")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# boxplot graphnabs  

age=[22, 25, 25, 30, 30, 30, 35, 40, 40, 45]

sns.boxplot(data=ages,color='lightgreen')

plt.title("age boxplot")
plt.show
```

```python
#dentisy graph like the cruncy graph

sns.kdeplot(ages,fill=False,color="orange")
plt.title("age density plot")
plt.show
```
